```python
# Creating a DataFrame from Dictionaries

# Dictionary with data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
# Creating a DataFrame from a CSV File

# Define the columns to select from the CSV file
selected_columns = ["Series_reference", "Period", "Data_value", "STATUS"]

# Read the CSV file and select specific columns
df_from_csv = pd.read_csv("nacgfs-jun-2023.csv", usecols=selected_columns)

# Display the first few rows to confirm the selection
df_from_csv.head()
```

```python
# Data Inspection Functions and their Meaning

# df.info(): Provides information about the DataFrame, including the number of rows, columns, data types, and any missing data.
df.info()

# df.head(n): Displays the first n rows of the DataFrame, useful for a quick overview.
df.head(5)

# df.describe(): Generates summary statistics for numerical columns, including count, mean, standard deviation, minimum, maximum, and quartiles.
df.describe()

# df.isnull().sum(): Returns the count of missing values in each column, helping identify data quality issues.
print("Missing values per column:")
df.isnull().sum()

# df.columns: Lists the column names in the DataFrame.
df.columns

# df.dtypes: Shows the data types for each column, useful for detecting inconsistencies.
df.dtypes
```

```python
# Data Exploration

# df["column"].value_counts(): Returns the count of unique values in a categorical column, helping understand the distribution.
age_counts = df["Age"].value_counts()
print("Age Counts:")
print(age_counts)

# df["column"].unique(): Lists all unique values in a column, useful for understanding diversity in the data.
unique_name = df["Name"].unique()
print("Unique Names:")
print(unique_name)
```
