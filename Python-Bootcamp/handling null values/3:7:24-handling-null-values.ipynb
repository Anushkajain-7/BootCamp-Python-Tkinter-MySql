```python
# End of Distribution
# It imputation is a method used to handle missing data where missing values are replaced with a value that lies at the end of the distribution of the variable.

# Quartiles and End-of-Distribution

# Quartiles divide a dataset into four equal parts:

# Q1 (First Quartile): The 25th percentile; represents the median of the lower half of the data.
# Q2 (Median): The 50th percentile; represents the middle value of the dataset.
# Q3 (Third Quartile): The 75th percentile; represents the median of the upper half of the data.
# These values can help identify the skewness of a distribution:

# If a distribution is symmetrical, the values of the quartiles will be evenly spaced.
# If a distribution is skewed, the quartiles will be unevenly spaced, indicating the direction of the skew.

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image_path = "Box Plot Explained.png"
img = mpimg.imread(image_path)
plt.imshow(img)
plt.axis('off')  
plt.title('Box Plot Explained')  
plt.show()
```

```python
# Calculate quartiles
q1 = df['Data_value'].quantile(0.25)  # First quartile (25th percentile)
q2 = df['Data_value'].quantile(0.5)   # Second quartile (median)
q3 = df['Data_value'].quantile(0.75)  # Third quartile (75th percentile)

print("First Quartile (Q1):", q1)
print("Median (Q2):", q2)
print("Third Quartile (Q3):", q3)

# Create a box plot to visualize skewness
sns.boxplot(x=df['Data_value'])
plt.title('Box Plot for Age')
plt.show()
```

```python
q1 = df['Data_value'].quantile(0.25)
q3 = df['Data_value'].quantile(0.75)
iqr = q3 - q1  # Interquartile Range

# lower and upper whiskers
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

print("Lower Whisker:", lower_whisker)
print("Upper Whisker:", upper_whisker)
```

```python
# Fill NaN with lower and upper whiskers
# Assuming one want to use these whiskers for imputation
df_lower_filled = df.fillna(lower_whisker)  # Fill with lower whisker
df_upper_filled = df.fillna(upper_whisker)  # Fill with upper whisker

# Display the results
print("Original DataFrame:")
df
```

```python
#BEFORE
# Calculate quartiles and visualize original distribution
q1 = df['Data_value'].quantile(0.25)  # First quartile (25th percentile)
q2 = df['Data_value'].quantile(0.5)   # Second quartile (median)
q3 = df['Data_value'].quantile(0.75)  # Third quartile (75th percentile)

print("First Quartile (Q1):", q1)
print("Median (Q2):", q2)
print("Third Quartile (Q3):", q3)

# Create a histogram to visualize original distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Data_value'], bins=30, color='lightblue', edgecolor='black')
plt.axvline(x=q1, color='red', linestyle='--', label='Q1 (25th percentile)')
plt.axvline(x=q2, color='green', linestyle='--', label='Q2 (50th percentile)')
plt.axvline(x=q3, color='blue', linestyle='--', label='Q3 (75th percentile)')
plt.title('Histogram of Data_value - Original Distribution')
plt.xlabel('Data_value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#AFTER
# Determine lower and upper whiskers for outlier detection
iqr = q3 - q1  # Interquartile Range
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

print("Lower Whisker:", lower_whisker)
print("Upper Whisker:", upper_whisker)

# Fill NaN values with lower and upper whiskers
df_lower_filled = df.fillna(lower_whisker)  # Fill with lower whisker
df_upper_filled = df.fillna(upper_whisker)  # Fill with upper whisker

# Create histograms to visualize the filled distributions
plt.figure(figsize=(14, 6))

# Histogram after filling with lower whisker
plt.subplot(1, 2, 1)
plt.hist(df_lower_filled['Data_value'], bins=30, color='orange', edgecolor='black')
plt.axvline(x=lower_whisker, color='red', linestyle='--', label='Lower Whisker')
plt.title('Histogram of Data_value - Filled with Lower Whisker')
plt.xlabel('Data_value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Histogram after filling with upper whisker
plt.subplot(1, 2, 2)
plt.hist(df_upper_filled['Data_value'], bins=30, color='green', edgecolor='black')
plt.axvline(x=upper_whisker, color='blue', linestyle='--', label='Upper Whisker')
plt.title('Histogram of Data_value - Filled with Upper Whisker')
plt.xlabel('Data_value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# KNN IMPUTATION

# Assuming df is your loan dataset loaded properly

# Select columns for KNN imputation
columns_for_knn = ['Period', 'Data_value']

# Visualize histograms before imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col].dropna(), bins=30, color='blue', edgecolor='black')
    plt.title(f'Histogram of {col} - Before Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.impute import KNNImputer

# Perform KNN imputation
imputer = KNNImputer(n_neighbors=5)
df_knn_imputed = pd.DataFrame(imputer.fit_transform(df[columns_for_knn]), columns=columns_for_knn)

# Visualize histograms after imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(df_knn_imputed[col], bins=30, color='green', edgecolor='black')
    plt.title(f'Histogram of {col} - After KNN Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# MICE (Multiple Imputation by Chained Equations)

# Visualize histograms before imputation (assuming df is your loan dataset)
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col].dropna(), bins=30, color='blue', edgecolor='black')
    plt.title(f'Histogram of {col} - Before Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
